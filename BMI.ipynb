```python
# Step 1: Install necessary libraries (if not already installed)
!pip install pandas numpy

# Step 2: Import libraries
import pandas as pd
import numpy as np

# Step 3: Define functions for BMI calculation and diet plan generation
def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def generate_diet_plan(bmi):
    if bmi < 18.5:
        return "Underweight: Increase caloric intake with nutrient-dense foods."
    elif 18.5 <= bmi < 24.9:
        return "Normal: Maintain a balanced diet with a variety of foods."
    elif 25 <= bmi < 29.9:
        return "Overweight: Focus on a lower-calorie, balanced diet with more vegetables and fruits."
    else:
        return "Obese: Follow a calorie-restricted diet with emphasis on lean proteins and vegetables."

# Step 4: Input data (you can also upload files for this)
weight = 70  # in kg
height = 1.75  # in meters

# Step 5: BMI Calculation
bmi = calculate_bmi(weight, height)
print(f"Your BMI is: {bmi}")

# Step 6: Generate and print the diet plan
diet_plan = generate_diet_plan(bmi)
print(f"Diet Plan: {diet_plan}")
```
